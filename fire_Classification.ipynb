```python
df1 = pd.read_csv('modis_2021_India.csv')
df2 = pd.read_csv('modis_2022_India.csv')
df3 = pd.read_csv('modis_2023_India.csv')
```

```python
# Check unique and n unique for all categorical features
for col in df.columns:
  if df[col].dtype == 'object':
    print(f"Column: {col}")
    print(f"Unique values: {df[col].unique()}")
    print(f"Number of unique values: {df[col].nunique()}")
    print("-" * 50)
```

```python
# Checking for missing values
print("Missing values in 2021 dataset:\n", df1.isnull().sum())
print("Missing values in 2022 dataset:\n", df2.isnull().sum())
print("Missing values in 2023 dataset:\n", df3.isnull().sum())

# Visualizing missing data (requires `missingno` library)
import missingno as msno
msno.matrix(df1, figsize=(10,4), color=(0.2, 0.4, 0.7))
plt.title("Missing values in df1 (2021)")
plt.show()
```

```python
# Check and drop duplicates
print("Duplicates before:", df1.duplicated().sum())
df1.drop_duplicates(inplace=True)
print("Duplicates after:", df1.duplicated().sum())
```

```python
# Heatmap of correlations
plt.figure(figsize=(10,6))
sns.heatmap(df1.drop(columns=['acq_date', 'satellite', 'instrument', 'daynight']).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap - 2021 Data")
plt.show()
```

```python
# Count plot for 'type'
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=df)
plt.title('Distribution of Fire Types')
plt.xlabel('Fire Type')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram of 'confidence'
plt.figure(figsize=(8, 6))
sns.histplot(df['confidence'], bins=20, kde=True)
plt.title('Distribution of Confidence')
plt.xlabel('Confidence')
plt.ylabel('Frequency')
plt.show()
```
